```python
from sklearn.datasets import load_boston
```

```python

buston=load_boston()
```

```python
buston
```

```python

print(buston.DESCR)
```

```python
Y=buston.target 
X=buston.data 
```

```python
X
```

```python
Y
```

## Crime

```python
import numpy as np
from sklearn.linear_model import LinearRegression
X_crime=X[:,0]
```

```python
model1=LinearRegression()
```

```python
X_crime=X_crime.reshape(-1,1)
```

```python
model1.fit(X_crime,Y)
```

```python
Y_pred1=model1.predict(X_crime)
```

```python
Y_pred1
```

```python
Y_pred1[:20]
```

```python
Y[:20]
```

```python
mae= np.abs(Y-Y_pred1).sum()/len(Y)
```

```python
mae
```
