## Pivot Assignment

```python
import pandas as pd
```

```python
dts = pd.DataFrame({'Consoles': ['PS5','XBOX-SeriesX','XBOX_ONE','XBOX360'],
                    'Gaming Titles': ['Call of Duty','Battlefield','NBA','Far Cry'],
                    'Price': [15000,5000,1000,8000]})
```

```python
dts
```

```python
pd.pivot_table(data=dts,index=['Consoles'],columns=['Gaming Titles'],values='Price',aggfunc='mean')
```

#### pivot tables with multiple argfuncs

```python

pd.pivot_table(data=dts,index=['Consoles'],
               values='Price',aggfunc=['sum','mean','count'],
              margins=True,margins_name='Grand Total')
```

## DataFrames Concatenation

```python
import numpy as np
```

```python
smartHub = {
    "Classes": ['Data Science', 'App Dev', 'Web Dev'],
    "Time": ['9-11', '11-2', '2-5']
}

itClass= pd.DataFrame(smartHub, index = 
                 ["Mon", "Wed", "Fri"])
itClass
```

```python
vertical_concat = pd.concat([dts, itClass], axis=0, sort=True)
vertical_concat
```

```python
Horizontal_concat = pd.concat([dts, itClass], axis=1, sort=False)
Horizontal_concat
```
