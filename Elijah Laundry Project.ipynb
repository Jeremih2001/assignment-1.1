```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from datetime import datetime

laundry= Tk()
laundry.title('Elijah')

screen_width=laundry.winfo_screenwidth()
screen_height=laundry.winfo_screenheight()
laundry.geometry(f'{screen_width}x{screen_height}')
laundry.config(bg='yellow')

k = {"singlet ":200, "bedspread ":500, "suit ":500,
     "curtain ":300, "rug ":1000, "wrappper ":400, "shirt ":200}


def click():
    save()
    showSelected()
    messagebox.OK
def save():
    
    name=namevar.get()
    phone=phonevar.get()
    sex=sexvar.get()
    cost=costvar.get()
    initialp=initialpvar.get()
    balance=balancevar.get()
    pdate=pdatevar.get()
    
    if len(name)==0:
        messagebox.showerror ('Name error!','Name field can\'t be empty')
        return
    
    if any(ch.isdigit() for ch in name):
        messagebox.showerror ('Error!','Name can\'t have numbers')
        return
    if len(name) <= 2:
        messagebox.showerror('Error!','Name is too short.')
        return
    if len(name) > 10:
        messagebox.showerror ('Error!','Name is too long.')

        return
    else:
        file = open('laundryrecord.txt','a')
        date = datetime.now()
        content=f'{name},{phone},{sex},{cost},{initialp},{balance},{pdate},{date}\n'
        file.write(content)
        file.close()
        
        messagebox.showinfo('Success!!','Success!\n Your Record has been taken ')
        

def Search():
    nam=namesvar.get()
    with open ('laundryrecord.txt','r') as file2:
        searchresult=file2.readlines()
    for user in searchresult:
        searchresult=user.split(',')[0]
        if searchresult==nam:
            print(searchresult)
            messagebox.showinfo('Success!!','Customer name found')
        

            return        
    file.close()
            

            
def show():
    name=namevar.get()
    l7.insert(END,name + '\n')


def login():
    global namesvar,namevar, phonevar, sexvar, costvar, initialpvar, balancevar, pdatevar, r, showSelected
    laundry2 = Toplevel()
    screen_width=laundry2.winfo_screenwidth()
    screen_height=laundry2.winfo_screenheight()
    laundry2.geometry(f'{screen_width}x{screen_height}')
    laundry2.config(bg='light cyan')
    
    hframe=Frame(laundry2)
    hframe.grid(row=0,column=0,columnspan=2)
    left=Frame(laundry2,bg='brown')
    left.grid(row=1,column=0)
    right=Frame(laundry2,bg='brown')
    right.grid(row=1,column=1,padx=5)
    
    
    Label(hframe,text='Laundry Inventory',font='courier 30 bold',bd=7,relief='raised',bg='#F8BD06',width=50).grid(row=0,column=0,columnspan=2)
   
    
    #right side window
    
    lh=LabelFrame(right,text='Display',bg='brown')
    lh.grid(row=1,column=1,padx=5)
    
    
    l1=Label(lh,text='Search',bg='brown',font = ('', 15))
    l1.grid(row=0,column=0,pady=1,padx=3)
    namesvar=StringVar()
    search=Entry(lh,textvariable=namesvar, font = ('', 15))
    search.grid(row=0,column=1,pady=3,padx=1)
    b2=ttk.Button(lh, text='Search', cursor='hand2',command=Search)
    b2.grid(row=0,column=2,pady=5)
    result=Text(lh,height = 15,width = 40,bg = "light cyan",)
    result.grid(row=2,column=0,columnspan=3,pady=3,padx=3)
    b1=ttk.Button(lh, text='View Entered Details', cursor='hand2',)
    b1.grid(row=3,column=2,pady=5)
    
    #left side window
    lb=LabelFrame(left,bg='red',text='fill the following details')
    lb.grid(row=1,column=0)
    
    l1=Label(lb,text='Customer Name ',bg='red',font = ('sans-serif', 15),).grid(row=1,column=1,pady=3,padx=3)
    namevar=StringVar()
    e1=Entry(lb,textvariable=namevar,font = ('', 15)).grid(row=1,column=2,pady=3,padx=3)
    
    l2=Label(lb,text=' Phone number ',bg='red',font = ('', 15)).grid(row=2,column=1,pady=3,padx=3)
    phonevar=StringVar()
    e2=Entry(lb,textvariable=phonevar,font = ('', 15)).grid(row=2,column=2,pady=3,padx=3)
    
    l3=Label(lb,text='     Sex     ',bg='red',font = ('', 15)).grid(row=3,column=1,pady=3,padx=3)
    sexvar=StringVar()
    sex=['Male','Female','other']
    combo=ttk.Combobox(lb,textvariable=sexvar,values=sex,font = ('', 15))
    combo.set('Other ')
    combo.grid(column=2,row=3,padx=3)
    
    
    l4=Label(lb,text='      Cost    ',bg='red',font = ('', 15)).grid(row=5,column=1,pady=3,padx=3)
    costvar=StringVar()
    e3=Entry(lb,textvariable=costvar,font = ('', 15)).grid(row=5,column=2,pady=3,padx=3)
    
    
    l5=Label(lb,text='  Initial payment  ',bg='red',font = ('', 15)).grid(row=6,column=1,pady=3,padx=3)
    initialpvar=StringVar()
    e4=Entry(lb,textvariable=initialpvar,font = ('', 15)).grid(row=6,column=2,pady=3,padx=3)
    
    
    l6=Label(lb,text='   Balance     ',bg='red',font = ('', 15)).grid(row=7,column=1,pady=3,padx=3)
    balancevar=StringVar()
    e5=Entry(lb,textvariable=balancevar,font = ('', 15)).grid(row=7,column=2,pady=3,padx=3)
    
       
    l7=Label(lb,text='Choose Cloth Type =>',bg='red',font = ('', 15)).grid(row=4,column=1)
    
    
    pdatevar=StringVar()
    cal=DateEntry(lb,textvariable=pdatevar,selectmode='Day',font = ('', 15))
    cal.grid(row=8,column=2,padx=3)
    
    cloth = Listbox(lb, cursor='hand2', selectmode = "multiple")
    cloth.grid( row=4,column=2)
    x = ["singlet " ,"bedspread " ,"suit " ,"curtain " ,"rug " ,"wrappper " ,"shirt "]
    for each_item in range(len(x)):
    
        cloth.insert(END, x[each_item])
        cloth.itemconfig(each_item,bg = "cyan" if each_item % 2 == 0 else "cyan")
        def showSelected():
            clothtype = []
            cname = cloth.curselection()
            for i in cname:
                op = cloth.get(i)
                clothtype.append(op)
            for val in clothtype:
                r=(val)
                v=k[r]
                print(r,k[r])
                l7.insert(END , k[r])
                
                print(v)
    
                
                

    l7=Entry(lb,text='  ',bg='red',font = ('', 15))
    l7.grid(row=9,column=1,pady=3,padx=3)

    ttk.Button(lb, text='save',command=click, cursor='hand2').grid(row=10,column=2,pady=5)
    ttk.Button(lb, text='test',command=showSelected, cursor='hand2').grid(row=10,column=1,pady=5)



l1=Frame(laundry,bg='red')
l1.pack()
b=Frame(laundry,bg='green')
b.pack(pady=30)
Label(l1,text='Welcome to my Laundry',font=("courier 40 bold"),bg='dark blue',relief='ridge').grid(column=50,row=1,padx=1,pady=1)
s=ttk.Style()
s.configure('my.TButton',font=('arial bold',15))
ttk.Button(b,text= 'Click \n  To\nLogin', cursor='hand2',style='my.TButton',command=login).grid(padx=5,pady=2)


laundry.mainloop()
```
